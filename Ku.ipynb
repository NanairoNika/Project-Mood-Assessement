```python
import csv
import nltk
from pymystem3 import Mystem
import pymorphy2
from itertools import groupby

from string import punctuation
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords

```

```python
nltk.download('stopwords')
nltk.download('punkt')
```

```python
filename = "CaP1.txt"
f = open(filename, encoding = 'utf8')
text = f.read()
```

```python
words = word_tokenize(text)
```

```python
morph = pymorphy2.MorphAnalyzer()
tokenized = []
for word in words:
    p = morph.parse(word)[0]
    tokenized.append(p.normal_form)

```

```python
stopWords = stopwords.words('russian')
tokenized = [token for token in tokenized if token not in stopWords\
            and token != " " \
            and token.strip() not in punctuation]
```

```python
cleaned = []
for word in tokenized:
    if (word != '»' and word != '«' and word != '—' and word != '..'):
        cleaned.append(word)
```

```python
preproctext = [word for word, _ in groupby(cleaned)]
```

```python
print (preproctext)
```

```python
filename = "dictionary.txt"
f = open(filename)
text = f.read()
```

```python
dicti = text.split()
```

```python
vecttext = []
i = 0
for el in dicti:
    counter = 0
    for word in preproctext:
        if el == word:
            counter = 1
            break
    vecttext.append(counter)
    i+=1
            
```

```python
reviews = []
stopWords = set(stopwords.words('english'))

with open("labeledTrainData.tsv", encoding='utf-8') as tsvfile:
    tsvreader = csv.reader(tsvfile, delimiter="\t")
    for line in tsvreader:
        data = line[1:]
        wordsFiltered = []
        words = word_tokenize(data)
        for w in words:
            if w not in stopWords:
                wordsFiltered.append(w)
        reviews.append(wordsFiltered)


print (reviews)
```
